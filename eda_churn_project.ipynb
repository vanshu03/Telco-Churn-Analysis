```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# importing data frame
df=pd.read_csv('Customer Churn.csv')
```

```python
df
```

```python
# inspection of data 
df.info()
```

```python
# since TotalCharges are object but it should be in float...
#objects are due to blank in TotalCharges
# Replace the Blank with either "NULL" or 0:
df["TotalCharges"]=df["TotalCharges"].replace(" ","0")
```

Converting its datatype

```python
df["TotalCharges"].astype("float")
```

# check null values

```python
df.isnull()
```

```python
#overall null value
df.isnull().sum().sum()
# no null value
```

```python
# for descriptive analysis 
df.describe()
```

```python
# duplicate checking
df.duplicated().sum()
#no duplicate value
```

# lets convert column(senior citizen) for easy to understand 

```python
df["SeniorCitizen"]= df["SeniorCitizen"].replace({0:"no",1:"yes"})
df["SeniorCitizen"]
```

```python
plt.figure(figsize=(4, 4))
ax = sns.countplot(x="Churn", data=df, hue="Churn", palette=["#708090", "#9370DB"], legend=True)
ax.bar_label(ax.containers[0],color='purple')
ax.bar_label(ax.containers[1],color='purple')
plt.title("Count of customers by Churns", fontsize=10,color="purple")
plt.xticks(color='purple', fontsize=12)  # Change x-axis tick label color
plt.yticks(color='grey', fontsize=12)  # Change y-axis tick label color
plt.show()
```

```python
plt.figure(figsize=(4,4))
gb=df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb["Churn"],labels=gb.index,autopct="%1.2f%%",shadow=1.4,colors=["#708090","#9370DB"],explode=[0.01,0.09], textprops={'color': 'purple', 'fontsize': 12})
plt.title("Percentage of Churned Customers",fontsize=10,color='purple')
plt.show()
```

```python
# from the given pie chart we can conclude that 26.54% of our customers have churned out
```

# now lets explore the reason behind it

```python
plt.figure(figsize=(5,4))
bx=sns.countplot(x="gender",data=df,hue="Churn",palette=["#708090", "#9370DB"], legend=True)
bx.bar_label(bx.containers[0],color="purple")
bx.bar_label(bx.containers[1],color="purple")
plt.title("Churn by Gender",fontsize=10,color="purple")
plt.xticks(color='purple', fontsize=12)  # Change x-axis tick label color
plt.yticks(color='grey', fontsize=12)  # Change y-axis tick label color
plt.xlabel('Gender',color="green")
plt.ylabel('Count',color="green")
plt.show()
```

```python
plt.figure(figsize=(4,4))
bx=sns.countplot(x="SeniorCitizen",data=df,hue="Churn",palette=["#708090", "#9370DB"], legend=True)
bx.bar_label(bx.containers[0],color="purple")
bx.bar_label(bx.containers[1],color="purple")
plt.title("Churn by SeniorCitizen",color="purple")
plt.xticks(color='purple', fontsize=12)  # Change x-axis tick label color
plt.yticks(color='grey', fontsize=12)  # Change y-axis tick label color
plt.xlabel('SeniorCitizen',color="green")
plt.ylabel('count',color="green")
plt.xticks(rotation=0,color="purple")
plt.yticks(color="grey")

plt.show()
```

#stack bar chart 

# comparatively the greater percentage of people in senior citizen category have churned

```python
total_counts = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack() * 100

# Plot
fig, ax = plt.subplots(figsize=(4, 4))  # Adjust figsize for better visualization

# Plot the bars
total_counts.plot(kind='bar', stacked=True, ax=ax, color=['#708090', '#9370DB'])  # Customize colors if desired

# Add percentage labels on the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width / 2, y + height / 2, f'{height:.1f}%', ha='center', va='center',color="purple")

plt.title('Churn by Senior Citizen (Stacked Bar Chart)',color="purple")
plt.xlabel('SeniorCitizen',color="green")
plt.ylabel('Percentage (%)',color="green")
plt.xticks(rotation=0,color="purple")
plt.yticks(color="grey")
plt.legend(title='Churn', bbox_to_anchor = (1,1))  # Customize legend locatio

plt.show()
```

```python
plt.figure(figsize=(5,4))
ax=sns.countplot(x="Contract",data=df,hue="Churn",palette=['#708090', '#9370DB'])
ax.bar_label(ax.containers[0],color="purple")
ax.bar_label(ax.containers[1],color="purple")
plt.title("Count of Customers by Contract",color="purple")
plt.xticks(color="purple")
plt.yticks(color="grey")
plt.xlabel('Contract',color="green")
plt.ylabel('Count',color="green")
plt.show()
#people who have month to month contract are likely to churn then from those who have 1 or 2 years of contract 
```

```python
# get name of columns to determine which characteristics are needed 
df.columns.values
```

```python
# from tenure to contract -'PhoneService', 'MultipleLines', 'InternetService','OnlineSecurity', 'OnlineBackup', 'DeviceProtection','TechSupport', 'StreamingTV', 'StreamingMovies', these 9 are needed to visualize 
# creating subplot 
# df contains the given columns
columns = [
    'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
    'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Create a subplot with 3 rows and 3 columns (since there are 9 columns in total)
fig, axes = plt.subplots(3, 3, figsize=(15, 15))

# Flatten axes array for easy iteration (2D to 1D)
axes = axes.flatten()

# Loop through the columns and plot a countplot for each
for i, column in enumerate(columns):
    bx=sns.countplot(data=df, x=column, ax=axes[i],hue="Churn",palette=['#708090', '#9370DB'])
    bx.bar_label(bx.containers[1],color="purple")
    bx.bar_label(bx.containers[0],color="purple")
    axes[i].set_title(f'Countplot for {column}',fontsize=11,color="purple")
    axes[i].set_xlabel(column,color="green")
    axes[i].set_ylabel('Count',color="green")
    axes[i].tick_params(axis='x', colors='purple')  
    axes[i].tick_params(axis='y', colors='grey') 

# Adjust the layout to avoid overlapping labels
plt.tight_layout()
plt.show()
```

#summary:

-Customers without additional services like Online Security, Online Backup, Tech Support, and Device Protection tend to have a higher churn rate.    
-Fiber optic internet service is associated with higher churn compared to DSL. 

-Streaming services (TV and Movies) show slightly reduced churn among users compared to non-users.
